## Total Revenue and Profits by Category

```python
##### The graph illustrates the total revenue and profits generated by each product category.
It highlights which categories are the most and least profitable, providing a clear comparison of financial performance
across different segments.
```

```python
## Best Peforming Product by Catergory
# Path to the SQLite database
path = r'C:\Users\calua\Desktop\3Signet\Sales_Adventure_Works.db'

# Connect to the SQLite database
conn = sqlite3.connect(path)

# SQL Query to retrieve top 10 performing products from 2015 and July 2016
query = """
SELECT
  CategoryName,
    SUM(s.OrderQuantity) AS Quantity,
	 SUM(s.OrderQuantity * p.ProductPrice) AS TotalSales,
	 sum(s.OrderQuantity * p.ProductCost) AS   totalcost,
	 (SUM(s.OrderQuantity * p.ProductPrice) -  sum(s.OrderQuantity * p.ProductCost)) as Profit
	  
	
FROM 
    AdventureWorks_Sales_Data AS s
INNER JOIN 
    AdventureWorks_Products AS p ON s.ProductKey = p.ProductKey
INNER JOIN 
    AdventureWorks_Product_Subcategories AS su ON p.ProductSubcategoryKey = su.ProductSubcategoryKey
INNER JOIN 
    AdventureWorks_Product_categories AS pc  ON pc.ProductCategoryKey = su.ProductCategoryKey
GROUP BY 
  CategoryName
ORDER BY 
   Quantity DESC

"""

# Read the data into a pandas DataFrame
df_product_Totals = pd.read_sql_query(query, conn)

# Close the connection
conn.close()
```

```python
# Define the data for the bar chart
categories = df_product_Totals['CategoryName']  # Categories
total_sales = df_product_Totals['TotalSales']   # Total Sales
profits = df_product_Totals['Profit']            # Profits

# Define the position of the bars
x = np.arange(len(categories))  # Label locations
width = 0.35  # Width of the bars

# Create the figure and the axes for Total Sales and Profits
fig, ax = plt.subplots(figsize=(12, 7))

# Bars for Total Sales
bars1 = ax.bar(x - width/2, total_sales, width, label='Total Sales ($)', color='#4CAF50', edgecolor='none')

# Bars for Profits
bars2 = ax.bar(x + width/2, profits, width, label='Profits ($)', color='#FF9800', edgecolor='none')

# Add labels, title, and custom x-axis tick labels
ax.set_xlabel('Category', fontsize=12, labelpad=15)
ax.set_ylabel('Values', fontsize=12, labelpad=15)
ax.set_title('Total Sales and Profits by Category', fontsize=16, weight='bold', pad=20)
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha='right', fontsize=11)

# Set y-axis limits with a specific minimum value
min_y_value = 10  # Set this to your desired minimum value
max_value = max(max(total_sales), max(profits))
buffer = max_value * 0.1  # 10% buffer above the max value
ax.set_ylim(min_y_value, max_value + buffer)  # Increase limit to max value plus buffer

# Annotate bars for Total Sales
for bar in bars1:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, f'${yval:,.2f}', ha='center', va='bottom', fontsize=10, color='black')

# Annotate bars for Profits with integer formatting
for bar in bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, f'${yval:,.2f}', ha='center', va='bottom', fontsize=10, color='black')

# Gridlines for the y-axis
ax.grid(True, which='major', axis='y', linestyle='--', alpha=0.7)

# Add a legend
ax.legend()

# Remove chart spines for a clean look
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Adjust layout for more space around labels and title
plt.tight_layout()

# Show the plot
plt.show()
```

## Top 10 Products by Revenue Performance Analysis and Sales trends

###### This analysis highlights the Top 10 Best Performing Products based on revenue.It provides insights into which products are generating the most income,
###### helping to identify key revenue drivers and potential areas for strategic focus .This also highlights the performance metrics and trends, as well as the
###### distribution of sales throughout the years.

```python
# Connect to the SQLite database
path = r'C:\Users\calua\Desktop\3Signet\Sales_Adventure_Works.db'
conn = sqlite3.connect(path)

# SQL Query to retrieve data over time
query = """
SELECT 
    strftime('%Y-%m', s.OrderDate) AS OrderMonth, 
    p.ProductName,
    SUM(s.OrderQuantity) AS Quantity,
    SUM(s.OrderQuantity * p.ProductPrice) AS TotalSales,
    SUM(s.OrderQuantity * p.ProductCost) AS TotalCost,
    (SUM(s.OrderQuantity * p.ProductPrice) - SUM(s.OrderQuantity * p.ProductCost)) AS Profit
FROM 
    AdventureWorks_Sales_Data AS s
INNER JOIN 
    AdventureWorks_Products AS p ON s.ProductKey = p.ProductKey
GROUP BY 
    OrderMonth, p.ProductName
ORDER BY 
    OrderMonth
    
"""

# Read the data into a pandas DataFrame
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Convert OrderMonth to datetime
df['OrderMonth'] = pd.to_datetime(df['OrderMonth'])

# Aggregate monthly totals
monthly_performance = df.groupby('OrderMonth').agg({'TotalSales': 'sum', 'Profit': 'sum', 'Quantity': 'sum'}).reset_index()

# Plotting Total Sales and Profit over time
plt.figure(figsize=(14, 7))
plt.plot(monthly_performance['OrderMonth'], monthly_performance['TotalSales'], label='Total Sales', marker='o', color='b')
plt.plot(monthly_performance['OrderMonth'], monthly_performance['Profit'], label='Total Profit', marker='o', color='g')
plt.title('Monthly Total Sales and Profit Over Time', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Amount ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Analyze top performing products
top_products = df.groupby('ProductName').agg({'TotalSales': 'sum', 'Profit': 'sum', 'Quantity': 'sum'}).reset_index()
top_products = top_products.sort_values(by='TotalSales', ascending=False).head(10)

# Visualization of Top Products
plt.figure(figsize=(12, 6))
plt.barh(top_products['ProductName'], top_products['TotalSales'], color='skyblue')
plt.xlabel('Total Sales ($)', fontsize=12)
plt.title('Top 10 Best Performing Products by Total Sales', fontsize=16)
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
# Convert OrderMonth to datetime
df['OrderMonth'] = pd.to_datetime(df['OrderMonth'])

# Calculate the total quantity for each month
monthly_trend = df.groupby('OrderMonth')['Quantity'].sum().reset_index()

# Define product life cycle stages
def assign_life_cycle_stage(quantity):
    if quantity == 0:
        return 'Decline'
    elif quantity < 50:
        return 'Introduction'
    elif 50 <= quantity < 200:
        return 'Growth'
    else:
        return 'Maturity'

# Assigning life cycle stages based on average monthly quantity 
df['LifeCycleStage'] = df.groupby('ProductName')['Quantity'].transform('mean').apply(assign_life_cycle_stage)

# Calculate the average quantity sold per life cycle stage for visualization
lifecycle_trend = df.groupby(['OrderMonth', 'LifeCycleStage'])['Quantity'].sum().unstack().fillna(0)

# Plotting the Monthly Trend Analysis on Quantity
plt.figure(figsize=(14, 7))
for stage in lifecycle_trend.columns:
    plt.plot(lifecycle_trend.index, lifecycle_trend[stage], marker='o', label=stage)

plt.title('Monthly Trend Analysis on Quantity by Product Life Cycle', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Quantity Sold', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Life Cycle Stage')
plt.tight_layout()
plt.show()
```

## Product Performance Analysis by Quantity and Territory

```python
# Path to the SQLite database
path = r'C:\Users\calua\Desktop\3Signet\Sales_Adventure_Works.db'

# Connect to the SQLite database
conn = sqlite3.connect(path)

# SQL Query to get product performance by territory and month
stacked_query = """
SELECT 
    p.ModelName,
    t.Country,
    strftime('%Y-%m', s.OrderDate) AS Month,  -- Extract year-month for grouping
    SUM(s.OrderQuantity) AS TotalQuantity,
    SUM(s.OrderQuantity * p.ProductPrice) AS TotalSales
FROM 
    AdventureWorks_Sales_Data AS s
INNER JOIN 
    AdventureWorks_Products AS p ON s.ProductKey = p.ProductKey
INNER JOIN 
    AdventureWorks_Territories AS t ON t.SalesTerritoryKey = s.TerritoryKey
GROUP BY 
    p.ModelName, t.Country, Month  -- Group by Month as well
ORDER BY 
    p.ModelName, Month;  -- Order by Product Model and Month
"""

# Read the data into a pandas DataFrame
df_stacked_Month = pd.read_sql_query(stacked_query, conn)

# Close the connection after reading the data
conn.close()


# Pivot the DataFrame to create a format suitable for a stacked bar chart
df_pivot = df_stacked_Month.pivot_table(index='Month', columns='ModelName', values='TotalQuantity', fill_value=0)

# Visualization: Stacked Bar Chart
df_pivot.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='tab10')
plt.title('Territory-Based Product Performance (Quantity Stacked)', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Quantity Sold', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Product Model', bbox_to_anchor=(1.05, 1), loc='upper left')  # Product models in the legend
plt.tight_layout()
plt.show()
```
